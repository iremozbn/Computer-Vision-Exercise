```python
import numpy as np
```

```python
img = np.zeros((512,512,3),np.uint8) #siyah bir resim oluşturur 512*512 boyutunda 3 ile de renk
img.shape
```

```python
plt.figure()
plt.imshow(img)
```

```python
cv2.line(img,(0,0),(200,512),(0,255,0),3) #başlama, durma, renk --çizgi oluşturduk
plt.figure()
plt.imshow(img)
```

```python
cv2.rectangle(img,(0,0),(256,256),(255,0,0), cv2.FILLED) #kare çizer
plt.figure()
plt.imshow(img)
```

```python
cv2.circle(img,(300,300),100,(0,0,255), cv2.FILLED) #daire çizer
plt.figure()
plt.imshow(img)
```

```python
cv2.putText(img,"IREM",(300,300), cv2.FONT_HERSHEY_COMPLEX,1,(255,0,0)) #yazı yazar
plt.figure()
plt.imshow(img)
```

```python
#threshold ekleme

img = cv2.imread("galata.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

(thresh,bfr) = cv2.threshold(img,20,255,cv2.THRESH_BINARY)
plt.figure()
plt.imshow(bfr)
```
